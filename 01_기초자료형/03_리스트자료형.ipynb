```python
# 리스트의 슬라이싱
a = [1, 2, 3, 4, 5]
print(a[0:3])
print(a[:3])
print(a[0:])
```

```python
# 리스트의 수정과 삭제 
a = [1, 2, 3]
a[2] = 4
print(a)
```

```python
# del 리스트 삭제하기 
a = [1, 2, 3]
del a[1]
print(a)
```

```python
a = [1, 2, 3, 4, 5]
del a[2:]
print(a)
```

```python
# 리스트 관련 함수 

# 1. 리스트에 요소 추가하기 - append 
a = [1, 2, 3]
a.append(4)
print(a)
```

```python
a.append([5, 6])
print(a)
```

```python
# 2. 리스트 정렬 - sort 
a = [ 5, 1, 4, 3, 2]
a.sort()
print(a)
```

```python
a.sort(reverse=True)
print(a)
```
